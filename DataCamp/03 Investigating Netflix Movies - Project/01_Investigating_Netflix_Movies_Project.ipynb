```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Start coding!
# Read in the CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Print the first five rows of the DataFrame
display(netflix_df.head())
```

```python
# Subset the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df.type == 'Movie']
```

```python
# Select only the columns of interest
netflix_movies = netflix_df_movies_only[["title", 'country', 'genre', 'release_year','duration']]
```

```python
# Print the first five rows of the new DataFrame
print(netflix_movies.head())
```

```python
# Create a scatter plot of duration versus year
plt.scatter(netflix_movies.release_year, netflix_movies.duration)

# Create a title
plt.title("Movie Duration by Year of Release")

# Show the plot
plt.show()
```

```python
colors = []

# Iterate over rows of netflix_movies_col_subset
for row, ser in netflix_movies.iterrows():
    if ser['genre'] == 'Children':
        colors.append('red')
    elif ser['genre'] == 'Documentaries' :
        colors.append('blue')
    elif ser['genre'] == "Stand-Up" :
        colors.append('green')
    else:
        colors.append('black')

# Inspect the first 10 values in your list        
print(colors[:11])
```

```python
# Set the figure style and initalize a new figure
fig = plt.figure(figsize=(12,8))
```

```python
# Create a scatter plot of duration versus release_year
plt.scatter(netflix_movies.release_year, netflix_movies.duration, c = colors)

# Create a title and axis labels
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")

# Show the plot
plt.show()
```
